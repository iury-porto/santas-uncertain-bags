```python
%matplotlib inline

import pandas as pd
import numpy as np
from scipy.stats import rv_continuous, norm, truncnorm, arcsine, chi2, gamma, triang, entropy
import matplotlib.pyplot as plt
from itertools import combinations_with_replacement
from time import time
```

```python
gift_types = ['horse', 'ball', 'bike', 'train', 'coal', 'book', 'doll', 'block', 'gloves']
gift_numbers = [1000, 1100, 500, 1000, 166, 1200, 1000, 1000, 200]
```

```python
convolutions = pd.read_csv('convolutions_1001_20_ready.csv')
```

```python
def max_number_of_packages(row, gift_types, gift_numbers):
    row_n = [row[i] for i in gift_types]
    max_packs = np.inf
    for rn,gn in zip(row_n,gift_numbers):
        if rn == 0:
            pass
        elif gn/rn < max_packs:
            max_packs = gn/rn
    return max_packs
```

```python
convolutions['max_packs'] = convolutions.apply(lambda row: max_number_of_packages(row, gift_types, gift_numbers), axis = 1).astype(int)
convolutions.to_csv('convolutions_1001_20_max.csv', index=False)
```
